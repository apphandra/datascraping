```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python

def web_driver():
    options = webdriver.ChromeOptions()
    options.add_argument("--verbose")
    options.add_argument('--no-sandbox')
    options.add_argument('--headless')
    options.add_argument('--disable-gpu')
    options.add_argument("--window-size=1920, 1200")
    options.add_argument('--disable-dev-shm-usage')
    driver = webdriver.Chrome(options=options)
    return driver
```

```python
driver = web_driver()

driver.get('https://onesolution.pertamina.com/Price')
```

```python
import pandas as pd
table =driver.find_element(By.XPATH,'//div[@class="table price"]')
rows = table.find_elements(By.TAG_NAME, 'tr')

header = table.find_element(By.TAG_NAME, 'thead')
header_cols = header.find_elements(By.TAG_NAME, 'th')
header_data = [col.text for col in header_cols]

# Loop through the rows and extract data
table_data = []
for row in rows:
    # Extract the columns in each row
    cols = row.find_elements(By.TAG_NAME, 'td')
    row_data = [col.text for col in cols]
    table_data.append(row_data)


# Print the extracted table data
df = pd.DataFrame(table_data, columns=header_data)

# Print the DataFrame
df
```
